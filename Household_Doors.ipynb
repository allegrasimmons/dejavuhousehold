## Match in Accuracy - Bathroom Door

```python
livingroombathroomdoor = [
    {'Audio': 'Bathroom Door  ', 'First': 'yes', 'Second': 'yes'},
    {'Audio': 'Bathroom Door 2', 'First': 'no', 'Second': 'no'},
    {'Audio': 'Bathroom Door 1', 'First': 'yes', 'Second': 'yes'}
]

livingroombathroomdoor = pd.DataFrame(livingroombathroomdoor)

# Calculate the matching percentage
matching_percentage = (livingroombathroomdoor['Second'] == 'yes').sum() / len(livingroombathroomdoor) * 100
mismatching_percentage = 100 - matching_percentage

# Create a bar chart for matching percentages
categories = ['Matching (Yes)', 'Mismatching (No)']
percentages = [matching_percentage, mismatching_percentage]
colors = ['lightgreen', 'lightcoral']

plt.figure(figsize=(8, 6))
plt.bar(categories, percentages, color=colors)
plt.ylabel('Percentage')
plt.title('Matching Percentages for Living Room Bathroom Door Second Result')

# Add labels to the bars
for i, percentage in enumerate(percentages):
    plt.annotate(f'{percentage:.1f}%', (i, percentage), ha='center', va='bottom')

plt.ylim(0, 110)  # Set the y-axis limit from 0% to 100%

plt.show()
```

# Summary Charts

```python
import pandas as pd
import matplotlib.pyplot as plt

# Living room data
livingroombathroomdoor = [
    {'Audio': 'Bathroom Door', 'First': 'yes', 'Second': 'yes'},
    {'Audio': 'Bathroom Door 2', 'First': 'no', 'Second': 'no'},
    {'Audio': 'Bathroom Door 1', 'First': 'yes', 'Second': 'yes'}
]
livingroombathroomdoor = pd.DataFrame(livingroombathroomdoor)

# Dining room data
diningroombathroomdoor = [
    {'Audio': 'Bathroom Door 2', 'First': 'no', 'Second': 'no'},
    {'Audio': 'Bathroom Door', 'First': 'yes', 'Second': 'yes'},
    {'Audio': 'Bathroom Door 1', 'First': 'no', 'Second': 'no'}
]
diningroombathroomdoor = pd.DataFrame(diningroombathroomdoor)

# Combine dataframes
combined_df = pd.concat([livingroombathroomdoor, diningroombathroomdoor], keys=['Living Room', 'Dining Room'])

# Count the occurrences of 'yes' and 'no' in each column for each room
counts_first = combined_df['First'].groupby(level=0).value_counts()
counts_second = combined_df['Second'].groupby(level=0).value_counts()

# Create bar charts for each room
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 12))

for idx, (room, ax) in enumerate(zip(['Living Room', 'Dining Room'], axes)):
    counts_first[room].plot(kind='bar', ax=ax, position=0, width=0.4, label='First')
    counts_second[room].plot(kind='bar', ax=ax, position=1, width=0.4, label='Second')

    ax.set_xticklabels(counts_first[room].index, rotation=0)
    ax.set_ylabel('Count')
    ax.set_title(f'Count of "yes" and "no" for Each Audio in {room}')

    ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Living room data
livingroombathroomdoor = [
    {'Audio': 'Bathroom Door', 'First': 'yes', 'Second': 'yes'},
    {'Audio': 'Bathroom Door 2', 'First': 'no', 'Second': 'no'},
    {'Audio': 'Bathroom Door 1', 'First': 'yes', 'Second': 'yes'}
]
livingroombathroomdoor = pd.DataFrame(livingroombathroomdoor)

# Dining room data
diningroombathroomdoor = [
    {'Audio': 'Bathroom Door 2', 'First': 'no', 'Second': 'no'},
    {'Audio': 'Bathroom Door', 'First': 'yes', 'Second': 'yes'},
    {'Audio': 'Bathroom Door 1', 'First': 'no', 'Second': 'no'}
]
diningroombathroomdoor = pd.DataFrame(diningroombathroomdoor)

combined_df = pd.concat([livingroombathroomdoor, diningroombathroomdoor], keys=['Living Room', 'Dining Room'])

match_counts = combined_df[combined_df == 'yes'].groupby(level=0).count()

# Calculate percentages of matches
total_audio = len(combined_df)
match_percentages = (match_counts / total_audio) * 100

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(match_percentages))
bar_width = 0.4

match_bars = ax.bar(x, match_percentages['First'], bar_width, label='Matched (First)', color='green')

ax.set_xticks(x)
ax.set_xticklabels(['Living Room', 'Dining Room'])
ax.set_ylabel('Percentage (%)')
ax.set_title('Percentage of Matches for Each Room')
ax.legend()

# Add percentages above the bars
def autolabel(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

autolabel(match_bars)

plt.tight_layout()
plt.show()
```
